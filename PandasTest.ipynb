```python
#pd.Series의 옵션이 무엇이 잇는지 확인
#index = > 할당값의 지정 이름, 중복 가능['a','a','b','c']
series = pd.Series([1,2,3,4], index=['a','b','c','d'],dtype=float)
series
```

```python
series = pd.Series([10,2,5,4], index=['a','b','c','d'],dtype=float)
series.sort_values(ascending=True)
```
