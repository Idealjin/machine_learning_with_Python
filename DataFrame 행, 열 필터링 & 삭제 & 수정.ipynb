```python
base_src = "./drive/MyDrive/machine_learning_data"
friend_src = base_src + "/friend.csv"
df = pd.read_csv(friend_src)
df.head()
```

```python
# index 2번에 해당하는 row 가져오기
  df.iloc[2]
```

```python
#column job에 해당되는 데이터 가져오기
df['job']
```

```python
#column job에 해당되는 데이터 가져오기(2)
df.loc[:,'job']
```

```python
# 슬라이싱 기능을 통해 여러 행 가져오기
df.iloc[[2,3]] #슬라이싱 개념을 쓰지 않음
```

```python
df.iloc[2:4]  #1~100의 경우 일일이 쓸수없어서 슬라이싱 사용
```

```python
df.iloc[:4] #3까지 가져와준다 0~3
```

```python
#조건 필터링 가져오기
# 30대 이상만 가져오기
df[df['age'] >= 30]
df[[True,True,True,True,False,True]] #이렇게 처리되는것과 같은 원리다.(False인것은 안보임)
```

```python
##job이 intern인 사람 가져오기
df[df['job'] == 'intern']
```

```python
## 조건이 여러개
## 30대 이상 40대 이하 (and Function)
df[(df['age'] >= 30) & (df['age'] <= 40)] 
```

```python
## 조건 여러개 30대 미만 혹은 40대 초과 (or function)
df[(df['age'] < 30) | (df['age']>40)]
```

```python
#in을 통한 포함 조건 걸기
#df[df['student','manager'] in df['job']]  <- 안됨
df[df['job'].apply(lambda x : x in ['student', 'manager'])] # x는 하나 검색하겠다는 뜻
```
